# seaborn

## categorical plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# basic categorical plot
# aggregates the numeric data and groups based on the category
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
# estimator parameter is to change the strategy of the aggregation
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
# let's analyze the distribution
# total bill vs the day of the week
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# how does smoking affect the distribution
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# violin plot = KDE plot + box plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# strip plot 
# but the plot does not give us much knowledge about the density
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
# let's use jitter to add some noise
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
# swarm plot shows density
# swarm = scatter + strip + violin plot
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
# factor plot
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```
