```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#heat maps, for heat maps to work effectively we need to have data in form of matrices
#matrix form can be achieved by either correlation or heatmaps

#so lets get correlation of the tips dataset

tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc,annot=True)
```

```python
#lets build pivot table for flights dataset

fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fp
```

```python
sns.heatmap(fp)
```

### cluster map

#### helps in clustering data and visualizing it

```python
sns.clustermap(fp)
```
