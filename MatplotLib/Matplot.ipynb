# matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# this will help us to make output plots visible in notebook
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) 
# or use plt.show() method, plt.show() is printing the plot
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('exponential func')
```

```python
# subplot args = rows,columns & plot number
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# this is the object oriented method of creating plot

fig = plt.figure() # figure object to playaround for graph

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height

axes.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Some func')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.2,0.3,0.4])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('smaller plot')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('x vs y')
axes[1].plot(y,x)
axes[1].set_title('y vs x')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green', linewidth = 5,alpha=0.5) # alpha is opacity
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green', linewidth = 5,linestyle='--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green', linewidth = 5,ls='steps') # linestyle or ls
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green', linewidth = 2,ls='-', marker='+', markersize=9)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green', linewidth = 2,ls='-', marker='o', 
        markersize=14, markerfacecolor='yellow',markeredgewidth=3,
        markeredgecolor='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green', linewidth = 2,ls='--')
ax.set_xlim([0,1]) #zoom into x axis
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green', linewidth = 2,ls='--')
ax.set_xlim([0,3]) #zoom into x axis
```
