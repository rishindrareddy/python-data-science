```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) # or use plt.show() method
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('exponential func')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# this is the object oriented method of creating plot

fig = plt.figure() # figure object to playaround for graph

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left,bottom,width,height

axes.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Some func')
```
